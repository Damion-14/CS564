```python
def maxValue(n):
    dp = [0]*(n+1)  # make room for dp[n]

    if n >= 2: dp[2] = 1
    if n >= 3: dp[3] = 2
    if n >= 4: dp[4] = 3

    for i in range(5, n+1):
        dp[i] = max(
            dp[i-2] + 1,
            dp[i-3] + 2,
            dp[i-4] + 3
        )

    return dp[n]

print(maxValue(11))
```
